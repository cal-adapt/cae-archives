# Determining Typical Meteorological Year

This notebook determines a typical meteorological year (TMY) for a given location, representing the mean weather conditions for a one-year period. A TMY are considered "typical" because the weather data is condensed into one year's worth of the most usual conditions. Because TMYs represent typical rather than extreme conditions, they are not suited for designing systems to meet the worst-case conditions occurring at a location.

In the Cal-Adapt: Analytics Engine, there are three options for investigating TMY:  
1. Absolute TMY: represents the historical reference period (1981-2010) or a future reference period based around a 30-year warming level time window. **Note: Absolute TMY data and display is not bias corrected, and should be used for visual data inspection only.**
2. Anomaly TMY: represents the difference between a designated future 30-year period and the historical baseline (1981-2010)
3. Extremes TMY: represents the difference between a designated extreme threshold and the historical baseline (1981-2010)

For any given location, we can calculate the desired TMY option and display:
1. Heatmap of TMY
2. Diurnal cycle of a typical day within the TMY

## example notebook to show capability to:
### (1) flexibly append metadata for a given transform 
### (2) show a progress bar and message as the transform is applied

```python
import xarray as xr
import export_helpers

ds = xr.open_dataset('ERA5-WRF_45km_monthly.nc',decode_coords=True)

# ds has no attributes so we will add some
# build a dictionary
# format: ds.attrs = {attribute name : attribute value}
var_att_dict = {'description' : 'TEMP at 2 M', 
                'units' : 'K', 
                'long_name' : 'T2',
                'output_time_step' : 'monthly',
                'output_type' : 'instantaneous',
                'coordinates' : 'y x'}

ds.T2.attrs = var_att_dict

ds.time.attrs = {'units' : 'months since 1950-09-01'}
ds.attrs = {'model' : 'Weather Research and Forecasting Model v4.1',
            'output_time_step' : 'monthly',
            'some_other_att' : 'some value',
            'history' : 'history 1'}

# now let's do the transform
ds_example_transform = export_helpers.example_transform(ds,'my_arg','my_next_arg',var='T2',time_bounds=['1951-01-01','2021-08-01'])
```

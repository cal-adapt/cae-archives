```python
import hvplot.pandas
import pandas as pd

import holoviews as hv
hv.extension("bokeh")

# Disable zoom / active scrolling
# Fixes the plot in place, better for website :) 
hv.plotting.bokeh.element.ElementPlot.active_tools = ['pan']
```

```python
df = pd.read_csv("tmy_los_angeles_international_airport_klax_wrf_miroc6_r1i1p1f1_mid-century.csv")
df["hour"] = range(1, len(df) + 1)
simulation = df["simulation"].iloc[0]

var_info = {
    "Air Temperature at 2m": {"label": "Air temperature at 2m", "units": "°C", "color": "#1f77b4"},
    "Dew point temperature": {"label": "Dew point temperature at 2m", "units": "°C", "color": "#ff7f0e"},
    "Relative humidity": {"label": "Relative humidity", "units": "0-100", "color": "#2ca02c"},
    "Instantaneous downwelling shortwave flux at bottom": {"label": "Instantaneous downwelling shortwave flux at bottom", "units": "W/m²", "color": "#d62728"},
    "Shortwave surface downward direct normal irradiance": {"label": "Shortwave surface downward direct normal irradiance", "units": "W/m²", "color": "#9467bd"},
    "Shortwave surface downward diffuse irradiance": {"label": "Shortwave surface downward diffuse irradiance", "units": "W/m²", "color": "#8c564b"},
    "Instantaneous downwelling longwave flux at bottom": {"label": "Instantaneous downwelling longwave flux at bottom", "units": "W/m²", "color": "#e377c2"},
    "Wind speed at 10m": {"label": "Wind speed at 10m", "units": "m/s", "color": "#7f7f7f"},
    "Wind direction at 10m": {"label": "Wind direction at 10m", "units": "0-360°", "color": "#bcbd22"},
    "Surface Pressure": {"label": "Surface pressure", "units": "Pa", "color": "#17becf"},
}

plots = []
for i, (col, info) in enumerate(var_info.items()):
    label = f"{info['label']} ({info['units']})"
    is_last = i == len(var_info) - 1
    p = df.hvplot(
        x="hour",
        y=col,
        width=750,
        height=175 if is_last else 150,
        color=info["color"],
        ylabel="", # No label on y-axis, since it's in the title
        label=label,
        legend="right",
        line_width=1,
        xlabel="Hour of year" if is_last else "",
    ).opts(show_legend=True, fontsize={"title": 12, "labels": 10, "ticks": 9}, xaxis="bottom" if is_last else None)
    plots.append(p)

layout = hv.Layout(plots).cols(1)

## Add suptitle 
#layout = layout.opts(hv.opts.Layout(title=f"Typical Meteorological Year (simulation: {simulation})"))

# Save the layout to an HTML file
hvplot.save(layout, "tmy_8760_by_variable.html")

# Display in-notebook
layout
```

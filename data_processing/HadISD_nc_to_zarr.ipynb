# Convert HadISD netcdf files to zarr stores
Import local HadISD netcdf files, and upload as zarr stores to the AWS bucket 
- Creation date: July 12, 2023<br> 
- Author: Nicole Keeney

```python
import xarray as xr
import os
```

```python
# Folder where the HadISD netcdf files are 
data_folder = "data/HadISD/"

# List all the files in that folder 
files = os.listdir(data_folder)

# Make sure there's not a rogue file in the directory. 
# We want netcdf files only 
files = [file for file in files if os.path.splitext(file)[1] == ".nc"]
```

```python
# Path to directory where you want to store the data on AWS
aws_path = "s3://cadcat/hadisd/"

# Loop through each file and add to directory 
for filename_nc in files: 
    # Open file using xarray 
    ds = xr.open_dataset(data_folder+filename_nc)
    
    # Path to zarr store in AWS bucket
    filepath_zarr = aws_path+filename_nc.split(".nc")[0]+".zarr"
    
    # Use xarray .to_zarr method to upload file as a zarr store to AWS bucket 
    ds.to_zarr(store=filepath_zarr, mode="w")
```

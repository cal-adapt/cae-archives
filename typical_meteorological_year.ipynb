# Determining Typical Meteorological Year

This notebook determines a typical meteorological year (TMY) for a given location, representing the median weather conditions for a one-year period. A TMY are considered "typical" because the weather data is condensed into one year's worth of the most usual conditions. Because TMYs represent typical rather than extreme conditions, they are not suited for designing systems to meet the worst-case conditions occurring at a location.

In the Cal-Adapt: Analytics Engine, a TMY represents the typical weather conditions at daily timescales over the entire available historical period of 1980 to 2014. TMY is calculated based off of the **Sandia (1970)** method commonly used by the NSRDB.
